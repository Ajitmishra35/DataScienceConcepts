Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

In pandas, a DataFrame is a two-dimensional, tabular data structure that consists of rows and columns. 
It is essentially a container for storing and manipulating structured data. 
You can think of it as a spreadsheet or a SQL table, where each column can be of a different data type.

On the other hand, a Series is a one-dimensional labeled array in pandas. 
It can be thought of as a single column of a DataFrame.
A DataFrame can be seen as a collection of Series where each Series represents a column.

```python
import pandas as pd


ages = pd.Series([25, 30, 27], name='Age')


data = {'Name': ['Alice', 'Bob', 'Claire'],
        'Age': [25, 30, 27],
        'Gender': ['Female', 'Male', 'Female']}

df = pd.DataFrame(data)


print("Series:")
print(ages)
print("\nDataFrame:")
print(df)
```

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

Among the options mentioned (Series, DataFrame, Panel), the DataFrame is mutable in nature.

DataFrame: DataFrames are mutable, meaning you can modify their content, add or remove columns, and perform various operations that change the structure or data within the DataFrame.

Series: Series, on the other hand, are mutable as well. You can modify the values within a Series, but you cannot change its size or shape. You can add, remove, or update elements in a Series.

Panel: The Panel class was used in earlier versions of pandas to handle three-dimensional data. However, it has been deprecated in recent versions, and you are encouraged to use the more flexible and memory-efficient MultiIndex DataFrames instead. Panels were mutable, but due to their deprecation, it's recommended to use DataFrames for similar functionality.

In summary, both Series and DataFrames are mutable, but the use of Panel is discouraged in favor of MultiIndex DataFrames.
